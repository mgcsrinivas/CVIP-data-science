```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize confirmed cases over time
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Country/Region', y='Confirmed', marker='o')
plt.xticks(rotation=90)
plt.title('Confirmed COVID-19 Cases by Country/Region')
plt.show()
```

```python
region_data = df.groupby('WHO Region')['Confirmed'].sum().sort_values(ascending=False)
print(region_data)
```

```python
correlation_matrix = df.corr()
print(correlation_matrix)
```

```python
df['Recovery Rate'] = (df['Recovered'] / df['Confirmed']) * 100
df['Death Rate'] = (df['Deaths'] / df['Confirmed']) * 100
df['Recovery Rate']
df['Death Rate']
```

```python
import matplotlib.pyplot as plt

countries = ["USA", "India", "China", "Brazil", "Russia"]
population = [331002651, 1380004385, 1444216107, 212559417, 145934462]

plt.bar(countries, population)
plt.xlabel('Countries')
plt.ylabel('Population')
plt.title('Population of Top 5 Countries')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
# Bar Chart - Top 10 Countries by Confirmed Cases
plt.figure(figsize=(10, 6))
plt.bar(df['Country/Region'], df['Confirmed'], color='blue')
plt.xlabel('Countries')
plt.ylabel('Confirmed Cases')
plt.title('Top 10 Countries by Confirmed Cases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
bar_width = 0.35
index = range(len(df))
plt.bar(index, df['Deaths / 100 Cases'], bar_width, label='Deaths / 100 Cases', color='red')
plt.bar([i + bar_width for i in index], df['Recovered / 100 Cases'], bar_width, label='Recovered / 100 Cases', color='green')
plt.xlabel('Countries')
plt.ylabel('Percentage')
plt.title('Deaths and Recovered per 100 Cases by Country')
plt.xticks([i + bar_width/2 for i in index], df['Country/Region'], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(df['WHO Region'].value_counts(), labels=df['WHO Region'].unique(), autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Cases by WHO Region')
plt.axis('equal')
plt.show()
```
